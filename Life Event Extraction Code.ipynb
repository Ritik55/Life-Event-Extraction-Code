```python
#list of all events
events= ['Wedding & Engagement','Relationship Begin','Anniversary','Relation End / Devorce','Graduation',
         'Admission','Exam','Research','Essay & Thesis','Job','Interview & Internship','Moving','Travel','Vacation',
         'Winning Award','Election / Promotion / Nomination','Publishing','Contract','song/ video/ album release','Acting',
         'Death','Give Birth','Illness','Surgery','Sports','Running','New Car','Lost Weight','Birthday','Lawsuit']

data= []
        
myfile = open("dataset.txt", "r")
myline = myfile.readline()
while myline:
    data.append(myline)
    myline = myfile.readline()
myfile.close()

dataset= []
for line in data:
    line = line.replace("\n", "")
    if line not in events:
        line = line.replace(" ", "")
    dataset.append(line)

# stemming data
from nltk.stem import PorterStemmer 
ps = PorterStemmer()
t1= []
for w in dataset:
    if w not in events:
        stem_w= ps.stem(w)
        stem_w = stem_w.replace(" ", "")
        stem_w.lower()
        t1.append(stem_w)
    else:
        t1.append(w)
dataset= t1

#print(dataset)

#creating dictionary for all events similar words
dic={}
for w in dataset:
    if w in events:
        word= w
        dic[w]= []
        v= w.lower()
        dic[w].append(v)
    else:
        dic[word].append(w)
```

```python
import nltk
# input tweet
tweet= input("Post a Tweet: ")

# input pre-processing
# tokenize
qt=nltk.word_tokenize(tweet)

#Remove Special characters
removetable=str.maketrans("", "", "'!@#$%^&*()_=-~’—‘”“\|][:';:,<.>/?`~")
qt=[x.translate(removetable) for x in qt]

#Decapitalized
qt=[element.lower() for element in qt]

tweet_list= []

ps = PorterStemmer()
t1= []
for w in qt:
    t1.append(ps.stem(w))
tweet_list= t1

print(tweet_list)
```
